# แฮช
จากบทที่แล้วเราได้เรียนรู้ทั้งการเรียงข้อมูลและการค้นหาข้อมูลทั้งแบบตามลำดับและแบบทวิภาคซึ่งมีข้อบ่งชี้ในการใช้งานที่ดีต่างกันตามสภาวะการใช้งาน ในบทนี้เราจะเรียนรู้วิธีที่จะสามารถค้นหาได้เร็วมากๆ ด้วยการใช้แฮช

## จุดประสงค์
* เพื่อให้เข้าใจถึงหลักการ การเขียนโปรแกรม สำหรับขั้นตอนวิธีการแฮช
* เพื่อให้เข้าใจถึงการออกแบบฟังก์ชันแฮช
* เพื่อให้สามารถแก้ไขการแฮชเมื่อเกิดการชนกันของข้อมูลได้

## 1. ตารางแฮช (hash table)
ตารางแฮชเป็นสิ่งที่เราจะใช้เก็บข้อมูล โดยปกติแล้วจะใช้ list โดยจะเก็บข้อมูลเข้าในแต่ละช่อง (slot) ซึ่งจริงๆ แล้วก็คือการอ้างด้วยดัชนีเหมือน list ทั่วไปแต่ช่องเป็นศัพท์เฉพาะที่ใช้กับตารางแฮช การอ้างอิงถึงข้อมูลในเชิงพื้นที่หน่วยความจำจะเหมือนดัชนีของ list คือเป็นช่องที่ 0, 1, 2 ต่อไปเรื่อยๆ โดยการเริ่มต้นตารางแฮชนั้นจะเริ่มที่ทุกสมาชิกว่างหมดหรือ None ถ้าเป็นภาษา java คือ Null ตัวอย่างดัง [รูปที่ 1](#figure_01) (จาก [interactivepython](http://interactivepython.org/runestone/static/pythonds/SortSearch/Hashing.html)) แสดงตัวอย่างตารางแฮชที่มีขนาด 11 โดยช่องจะเริ่มที่ 0 และสิ้นสุดที่ 10

## 2. ฟังก์ชันแฮช (hash function)
เมื่อเราต้องการใส่ข้อมูลเข้าไปในตารางแฮชสิ่งที่เราต้องระบุคือช่อง แต่การที่จะบอกได้ว่าเราควรจะใส่ข้อมูล item ไว้ที่ช่องไหนต้องผ่านการคำนวณฟังก์ชันก่อน ซึ่งเราเรียกว่าฟังก์ชันแฮช รูปแบบที่ง่ายที่สุดของฟังก์ชันแฮชคือ การหารแบบเอาเศษ (modulo) ด้วยขนาดของตารางแฮช จะขอเรียกสั้นๆ ว่า mod และขอเรียกผลการคำนวณฟังก์ชันแฮชว่า ค่าแฮช (hash value) เราสามารถสรุปให้อยู่ในรูปฟังก์ชันทางคณิตศาสตร์ได้คือ $h(item) = item \% 11$ ยกตัวอย่าง [ตารางที่ 1](#table_01) แสดงค่า item ที่ต้องการนำเข้าตารางแฮช ค่าแฮชคือค่าที่ได้จากการคำนวณฟังก์ชันแฮช

### 2.1 วิธีการพับ (folding method)
เป็นวิธีการสร้างฟังก์ชันแฮชโดยการแบ่งข้อมูลเป็นจำนวนหลักย่อยๆ เท่าๆ กัน แล้วจึงนำหลักย่อยๆ นั้นมาบวกกัน เช่น 436-555-4601 เป็นหมายเลขโทรศัพท์ 10 หลัก วิธีที่ง่ายที่สุดตั้งแต่การออกแบบตารางแฮชคือ สร้าง list ขนาด $10^{10}$ เพื่อเก็บหมายเลขโทรศํพท์ทั้งหมดซึ่งจะไม่เกิดการชนแน่นอน แต่หากเราต้องการเก็บข้อมูลแค่ไม่กี่สิบตัว จะเห็นว่ามีการใช้เนื้อที่ที่ไม่มีความจำเป็น ดังนั้นเราจึงอยากสร้าง list ให้มีขนาดเท่าๆ กับจำนวนข้อมูลที่ต้องการจะเก็บ ข้อมูลที่เป็นตัวเลขยาวๆ นี้มีอีกตัวอย่างหนึ่งคือเลขบัตรประจำตัวประชาชน ได้มีผู้เสนอวิธีการคำนวณฟังก์ชันแฮชกับตัวเลขยาวๆ ด้วยการแบ่งเป็นกลุ่มกลุ่มละ 2 หลัก ได้เป็น (43,65,55,46,01) แล้วนำค่าในแต่ละกลุ่มมารวมกัน $43+65+55+46+01 = 210$ แต่ทุกครั้งต้องมีการ mod ด้วยขนาดของตาราง เรากำหนดไว้เป็น 11 ดังนั้นในกรณีนี้เราจะได้ค่าแฮชคือ $210\%11 = 1$

### 2.3 วิธีการสำหรับ string

## 4. การสร้าง HashTable
เราจะสร้างโครงสร้างข้อมูลที่มีลักษณะคล้ายกับ dictionary ของ python โดยข้อมูลที่สำคัญมีอยู่ 2 ตัวคือ key และ value โดยเราจะตั้งชื่อว่า HashTable

### 4.1 ADT ของ HashTable
ขั้นแรกเราจะต้องกำหนดการทำงานและเมธอดสำหรับ HashTable ก่อน ซึ่งประกอบด้วย
* HashTable() สร้าง HashTable ว่างๆ แล้วคืนค่า HashTable
* put(key, value) เพิ่มค่าเข้าใน HashTable ด้วยคู่ key กับ value หาก key มีอยู่ใน HashTable อยู่แล้วให้ทำการแทนที่ value ลงไป
* get(key) ค้นหา key แล้วคืนค่าเป็น value ของ key นั้น หากไม่เจอ key ให้คืนค่า None แทน
* del ลบค่า key และ value โดยผ่านคำสั่ง del HashTable[key]
* len คืนค่าจำนวนข้อมูลใน HashTable
* in ตรวจสอบว่ามี key อยู่ใน HashTable หรือไม่ผ่านคำสั่ง key in HashTable โดยจะคืนค่า True ในกรณีที่มีหรือ False ในกรณีที่ไม่มี

### 4.3 การสร้าง HashTable
มีรายละเอียดรหัสดังนี้

```python
class HashTable:
    def __init__(self):
        self.size = 11
        self.slots = [None] * self.size
        self.data = [None] * self.size

    def put(self,key,data):
      hashvalue = self.hashfunction(key,len(self.slots))

      if self.slots[hashvalue] == None:
        self.slots[hashvalue] = key
        self.data[hashvalue] = data
      else:
        if self.slots[hashvalue] == key:
          self.data[hashvalue] = data  #replace
        else:
          nextslot = self.rehash(hashvalue,len(self.slots))
          while self.slots[nextslot] != None and \
                          self.slots[nextslot] != key:
            nextslot = self.rehash(nextslot,len(self.slots))

          if self.slots[nextslot] == None:
            self.slots[nextslot]=key
            self.data[nextslot]=data
          else:
            self.data[nextslot] = data #replace

    def hashfunction(self,key,size):
         return key%size

    def rehash(self,oldhash,size):
        return (oldhash+1)%size

    def get(self,key):
      startslot = self.hashfunction(key,len(self.slots))

      data = None
      stop = False
      found = False
      position = startslot
      while self.slots[position] != None and  \
                           not found and not stop:
         if self.slots[position] == key:
           found = True
           data = self.data[position]
         else:
           position=self.rehash(position,len(self.slots))
           if position == startslot:
               stop = True
      return data

    def __getitem__(self,key):
        return self.get(key)

    def __setitem__(self,key,data):
        self.put(key,data)
```

บรรทัดที่ 2-5 เป็นคอนสตรัคเตอร์มีจุดสำคัญคือการกำหนดขนาดของ HashTable <br>
บรรทัดที่ 7-26 เป็นเมธอด put ซึ่งจะต้องการคำนวณค่าแฮชก่อนในบรรทัดที่ 8 เพื่อนำมาตรวจสอบว่าว่างหรือไม่ หากว่างก็สามารถใส่ข้อมูลได้เลย แต่หากไม่ว่างจะต้องตรวจสอบก่อนว่าเป็น key เดียวกันเพื่อแทนค่าหรือ key ใหม่ที่เกิดการชนในบรรทัดที่ 14 หากเป็นการชนจะต้องมีการแก้การชน <br>
บรรทัดที่ 31-32 เป็นเมธอด rehash ใช้ในการแก้การชน ซึ่งในรหัสนี้ใช้วิธีการตรวจสอบเชิงเส้นที่ทำการขยับทีละช้อง <br>
บรรทัดที่ 34-50 เป็นเมธอด get ซึ่งมีการทำงานคล้ายๆ กับ put เพราะการค้นหาจะต้องตามเส้นทางการหาช่องแบบเดียวกันกับตอนค้นหาช่องว่างเพื่อใส่ข้อมูล <br>
\__getitem\__ เป็นเมธอดที่ถูกโอเวอร์ไรด์เพื่อใช้ในการดึงค่าข้อมูลด้วย [ ] <br>
\__setitem\__ เป็นเมธอดที่ถูกโอเวอร์ไรด์เพื่อใช้ในการกำหนดข้อมูลด้วย [ ]
