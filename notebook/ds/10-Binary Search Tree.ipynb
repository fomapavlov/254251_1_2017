# ต้นไม้ค้นหาแบบทวิภาค
ต้นไม้ค้นหาแบบทวิภาค (binary search tree) เป็นโครงสร้างข้อมูลที่ถูกออกแบบมาเพื่อใช้ในการค้นหาอย่างมีประสิทธิภาพ ซึ่งจะแตกต่างจากต้นไม้แบบทวิภาคปกติที่ไม่ได้ออกแบบมาเพื่อใช้ในการค้นหา รวมถึงฮีพซึ่งออกแบบมาเพื่อให้คืนค่าที่มีความสำคัญมากที่สุดให้เร็วที่สุดเท่านั้น

## จุดประสงค์
* เพื่อให้เข้าใจถึงคุณสมบัติของต้นไม้ค้นหาแบบทวิภาค
* เพื่อให้สามารถสร้างโครงสร้างข้อมูลต้นไม้ค้นหาแบบทวิภาคได้

## 1. ADT ของ BST
เราจะออกแบบการใช้งานของต้นไม้ (ขอเรียกสั้นๆ แทนต้นไม้ค้นหาแบบทวิภาค) โดยกำหนดชื่อคลาสว่า BST โดยให้มีการเก็บ key และ value หรือ payload คล้ายๆ กับ dictionary โดย key จะถูกใช้ในการค้นหาเพื่อดึง payload มาใช้งาน ADT ของ BST มีดังนี้
* BST() สร้างต้นไม้ว่างๆ
* put(key,value) เพิ่มคู่ลำดับ key และ value เข้าในต้นไม้ หากมี key อยู่แล้วให้ทำการแทนที่ value ไป
* get(key) คืนค่า value ที่สอดคล้องกับ key หากไม่มี key ให้คืนค่า None แทน
* del ลบค่า key และ value โดยผ่านคำสั่ง del BST[key]
* len คืนค่าจำนวนข้อมูลในต้นไม้
* in ตรวจสอบว่ามี key อยู่ในต้นไม้หรือไม่ผ่านคำสั่ง key in BST โดยจะคืนค่า True ในกรณีที่มีหรือ False ในกรณีที่ไม่มี

## 2. คุณสมบัติของ BST 
คุณสมบัติของ BST คือ ที่ปมใดๆ key ของปมนั้นจะต้องมีค่ามากกว่าทุกปมในต้นไม้ย่อยซ้าย และจะต้องมีค่าน้อยกว่าทุกปมในต้นไม้ย่อยขวา [รูปที่ 1](#figure_01) แสดงลำดับของการสร้างต้นไม้ด้วยข้อมูล 70,31,93,94,14,23,73 ตามลำดับ มีขั้นตอนดังนี้
* เริ่มจากการเพิ่ม 70 เป็นรากของต้นไม้ 
* 31 มีค่าน้อยกว่า 70 ดังนั้นจึงกลายเป็นลูกซ้ายของ 70
* 93 มีค่ามากกว่า 70 ดังนั้นจึงกลายเป็นลูกขวาของ 70
* 94 มีค่ามากกว่า 70 ดังนั้นจึงควรจะกลายเป็นลูกขวาของ 70 แต่ 70 มีลูกขวาคือ 93 อยู่แล้ว จึงเปรียบเทียบกับ 93 พบว่ามีค่ามากกว่า 93 ดังนั้นจึงกลายเป็นลูกขวาของ 93
* 14 มีค่าน้อยกว่า 70 ดังนั้นจึงควรจะกลายเป็นลูกซ้ายของ 70 แต่ 70 มีลูกซ้ายคือ 31 อยู่แล้ว จึงเปรียบเทียบกับ 31 พบว่ามีค่าน้อยกว่า 31 ดังนั้นจึงกลายเป็นลูกซ้ายของ 31
* 23 และ 73 ก็ทำการเปรียบเทียบเพื่อเพิ่มข้อมูลแบบเดียวกัน

## 3. การสร้าง BST
เราจะใช้วิธีการอ้างอิงปมในการสร้างต้นไม้ ซึ่งจะสร้างคลาสขึ้นมา 2 คลาสคือ TreeNode กับ BST โดย TreeNode จะเป็นคลาสที่เก็บข้อมูลของปมไว้ ในขณะที่ BST จะเป็นคลาสที่เก็บเฉพาะรากเป็นหลักและมีเมธอดในการจัดการต้นไม้

### 3.1 คลาสของ TreeNode
คลาสของ TreeNode ประกอบไปด้วยเมธอดที่จะถูกใช้ใน BST ในส่วนของคอนสตรัคเตอร์จะเป็นการสร้างตัวแปรและกำหนดค่า 

```python
class TreeNode:
   def __init__(self,key,val,left=None,right=None,
                                       parent=None):
        self.key = key
        self.payload = val
        self.leftChild = left
        self.rightChild = right
        self.parent = parent
```

มีประเด็นที่น่าสนใจในการเขียนภาษาไพทอน คือการรับพารามิเตอร์แล้วกำหนดค่าให้ในบรรทัดที่ 2 เช่น left=None การกำหนดค่าในลักษณะนี้หมายความว่า เราเลือกได้ว่าจะส่งพารามิเตอร์left หรือไม่ หากไม่ส่งจะเหมือนส่งค่า None มา

ลองดูตัวอย่างอื่นง่ายๆ 

```python
def myFunction(a,b=1):
    return a+b

print(myFunction(2,2))
print(myFunction(2))
```

จากตัวอย่างเราจะเห็นได้ว่าหากเราไม่ส่งค่าผ่าน b ในบรรทัดที่ 5 ค่า b จะถูกกำหนดเป็น 1 โดยอัตโนมัติ

กลับมาที่ต้นไม้ของเรา TreeNode ควรจะมีเมธอดชวยเหลืออื่นๆ อีกเช่น การถามว่ามีลูกซ้ายลูกขวาหรือไม่ หรือตัวมันเองเป็นลูกซ้ายลูกขวาของใครหรือไม่ รวมถึงการเปลี่ยนแปลงค่าในปม สรุปรวมรหัสได้ดังนี้

```python
class TreeNode:
    def __init__(self,key,val,left=None,right=None,
                                       parent=None):
        self.key = key
        self.payload = val
        self.leftChild = left
        self.rightChild = right
        self.parent = parent

    def hasLeftChild(self):
        return self.leftChild

    def hasRightChild(self):
        return self.rightChild

    def isLeftChild(self):
        return self.parent and self.parent.leftChild == self

    def isRightChild(self):
        return self.parent and self.parent.rightChild == self

    def isRoot(self):
        return not self.parent

    def isLeaf(self):
        return not (self.rightChild or self.leftChild)

    def hasAnyChildren(self):
        return self.rightChild or self.leftChild

    def hasBothChildren(self):
        return self.rightChild and self.leftChild

    def replaceNodeData(self,key,value,lc,rc):
        self.key = key
        self.payload = value
        self.leftChild = lc
        self.rightChild = rc
        if self.hasLeftChild():
            self.leftChild.parent = self
        if self.hasRightChild():
            self.rightChild.parent = self
```

โดยแต่ละเมธอดมีการทำงานคือ (ซ้ายกับขวามีการทำงานคล้ายๆ กัน)
* hasLeftChild ทำการคืนค่าว่ามีลูกซ้ายหรือไม่ เราสามารถแยกแยะว่ามีหรือไม่มีได้ด้วยการเป็น None หรือไม่
* isLeftChild ทำการตรวจสอบว่าเป็นลูกซ้ายของพ่อของมันหรือไม่ ซึ่งต้องมีการตรวจสอบการอ้างอิงกับปมที่เป็นพ่อ
* isRoot ทำการตรวจสอบว่าเป็นรากหรือไม่ ทำได้โดยการตรวจสอบว่ามีพ่อหรือไม่
* isLeaf ทำการตรวจสอบว่าเป็นใบหรือไม่ ทำได้โดยการตรวจสอบว่าไม่มีลูก
* hasAnyChildren ทำการตรวจสอบว่ามีลูกหรือไม่
* hasBothChildren ทำการตรวจสอบว่ามีทั้งลูกซ้ายและลูกขวาหรือไม่
* replaceNodeData ทำการเปลี่ยนแปลงค่าในปม มีประเด็นที่สำคัญคือ เมื่อมีการเปลี่ยนแปลงลูกซ้ายหรือขวาจะต้องกำหนดให้พ่อของลูกนั้นๆ เป็นปมนั้นเองด้วยในบรรทัดที่ 39-42

### 3.2 คลาสของ BST
ส่วนคอนสตรัคเตอร์จะสร้างตัวแปรเพื่อเก็บรากของต้นไม้เป็นหลัก ที่เก็บแค่รากก็เพราะว่าโครงสร้างต้นไม้เป็นแบบอ้างอิง เราสามารถอ้างอิงจากรากเพื่อไปสู่ปมใดๆ ในต้นไม้ได้

```python
class BST:

    def __init__(self):
        self.root = None
        self.size = 0

    def length(self):
        return self.size

    def __len__(self):
        return self.size

    def __iter__(self):
        return self.root.__iter__()
```

การเพิ่มข้อมูลด้วยเมธอดput มีการทำงานคือ ถ้าต้นไม้ว่างจะทำการสร้างรากขึ้นมาแล้วกำหนดค่าเลย แต่ถ้าต้นไม่ไม่ว่างจะทำการค้นหาตำแหน่งที่ key ควรจะอยู่ แต่หากว่าเจอ key อยู่แล้วจะทำการแทนค่า value เข้าไปแทน ซึ่งกระบวนการค้นหานั้นเราจะใช้ความสัมพันธ์แบบเวียนเกิด แต่ในกรณีต้นไม้ว่างเราไม่ต้องทำดังนั้นเราจึงจะสร้างเมธอดชื่อ \_put ขึ้นมาเพื่อใข้ในความสัมพันธ์แบบเวียนเกิด

```python
class BST:

    def __init__(self):
        self.root = None
        self.size = 0

    def length(self):
        return self.size

    def __len__(self):
        return self.size

    def __iter__(self):
        return self.root.__iter__()
    
    def put(self,key,val):
        if self.root:
            self._put(key,val,self.root)
        else:
            self.root = TreeNode(key,val)
        self.size = self.size + 1

    def _put(self,key,val,currentNode):
        if key < currentNode.key:
            if currentNode.hasLeftChild():
                   self._put(key,val,currentNode.leftChild)
            else:
                   currentNode.leftChild = TreeNode(key,val,parent=currentNode)
        else:
            if currentNode.hasRightChild():
                   self._put(key,val,currentNode.rightChild)
            else:
                   currentNode.rightChild = TreeNode(key,val,parent=currentNode)
                    
    def __setitem__(self,k,v):
        self.put(k,v)
```

ขั้นตอนการทำงานของ \_put มีดังนี้คือ
* เริ่มที่รากโดยมองว่าเป็นปมหนึ่งปม ทำการค้นหาด้วยการเปรียบเทียบ key ใหม่ กับ key ที่ปม หาก key ใหม่มีค่าน้อยกว่า ให้ค้นหาแบบเดียวกันที่ลูกด้านซ้าย หากมีค่ามากกว่า ให้ค้นหาแบบเดียวกันที่ลูกด้านขวา การเปรียบเทียบอยู่ในบรรทัดที่ 24 และการค้นหาแบบเดียวกันคือการเรียกความสัมพันธ์เวียนเกิดในบรรทัดที่ 26 และ 31
* หากพบว่าต้นการค้นหาที่ลูกซ้าย แต่ลูกซ้ายไม่มี นั่นคือเจอที่สำหรับ key ใหม่แล้วให้สร้าง TreeNode ขึ้นมาเป็นปมใหม่ ในกรณีลูกขวาก็เช่นกัน การตรวจสอบว่ามีลูกซ้ายหรือลูกขวาหรือไม่อยู่ในบรรทัดที่ 25 และ 30 

และด้วยการสร้างเมธอดput ในลักษณะนี้ทำให้เราสามารถใช้การกำหนดค่าได้ด้วยการโอเวอร์ไรด์เมธอด\_\_setitem\_\_ ในบรรทัดที่ 35-36

การดึงข้อมูลด้วย get มีการทำงานคล้ายๆ กับ put โดยทำการค้นหาด้วยความสัมพันธ์เวียนเกิดผ่าน \_get ซึ่งทำการค้นหาด้วยการเปรียบเทียบเช่นเดียวกับ \_put แต่จะเพิ่มในกรณีที่หาไม่เจอซึ่งก็คือการค้นหาไปจนเจอการอ้างอิงปมที่เป็น None และในกรณีที่หาเจอจะคืนค่าการอ้างอิงปมในบรรทัดที่ 12-15 เพื่อให้ get นำปมนั้นไปคืนค่า payload แต่หากการอ้างอิงปมเป็น None จะคืนค่า None ในบรรทัดที่ 4-7 

```python
def get(self,key):
    if self.root:
        res = self._get(key,self.root)
        if res:
               return res.payload
        else:
               return None
    else:
        return None

def _get(self,key,currentNode):
    if not currentNode:
        return None
    elif currentNode.key == key:
        return currentNode
    elif key < currentNode.key:
        return self._get(key,currentNode.leftChild)
    else:
        return self._get(key,currentNode.rightChild)

def __getitem__(self,key):
    return self.get(key)
```

และด้วยการเพิ่มเมธอด\_get ทำให้เราสามารถใช้ในการสร้างคำสั่ง in เพื่อตรวจสอบว่ามี key อยูในต้นไม้หรือไม่ด้วยการโอเวอร์ไรด์ \_\_contains\_\_

```python
def __contains__(self,key):
    if self._get(key,self.root):
        return True
    else:
        return False
```

ส่วนที่ยากที่สุดคือการลบปม ซึ่งสิ่งแรกที่ต้องทำคือการค้นหาปมที่ต้องการจะลบออกก่อนโดยใช้ \_get ในการใช้งานเราสามารถใช้คำสั่ง del ได้โดยการโอเวอร์ไรด์ \_\_delitem\_\_ ดังรหัส

```python
def delete(self,key):
   if self.size > 1:
      nodeToRemove = self._get(key,self.root)
      if nodeToRemove:
          self.remove(nodeToRemove)
          self.size = self.size-1
      else:
          raise KeyError('Error, key not in tree')
   elif self.size == 1 and self.root.key == key:
      self.root = None
      self.size = self.size - 1
   else:
      raise KeyError('Error, key not in tree')

def __delitem__(self,key):
    self.delete(key)
```

จากรหัสจะเห็นได้ว่า delete ยังเป็นเพียงแค่เปลือกห่อหุ้มการลบปมซึ่งจะเกิดขึ้นในเมธอด remove ซึ่งสิ่งที่เราต้องคำนึงถึงคือเมื่อลบปมออกแล้วจะทำอย่างไรกับพ่อและลูกของปมนั้น หลักๆ คือการพิจารณาที่ลูก โดยแบ่งได้เป็น 3 กรณีคือ
1. ไม่มีลูก
2. มีลูกเดียว
3. มีลูกสอง

สำหรับกรณีที่ไม่มีลูกนั้นหมายความว่าตัวมันเองเป็นใบดังนั้นสิ่งที่ต้องทำก็เพียงแค่เปลี่ยนการอ้างอิงของพ่อของมันสู่ตัวมันให้เป็น None ดัง [รูปที่ 2](#figure_02) (จาก [interactivepython](http://interactivepython.org/runestone/static/pythonds/Trees/SearchTreeImplementation.html))

กรณีที่สองที่มีลูกเพียงตัวเดียวคือลูกซ้ายให้ทำการนำลูกมาแทนที่ ซึ่งมีกรณีย่อยๆ ที่ต้องคำนึงถึงคือ
1. หากเป็นปมลูกซ้าย ให้ทำการเปลี่ยนแปลงการอ้างอิงลูกซ้ายของพ่อของมันไปยังลูกซ้ายของมัน และเปลี่ยนแปลงการอ้างอิงพ่อของลูกซ้ายของมันไปยังพ่อของมัน
2. หากเป็นปมลูกขวา ให้ทำการเปลี่ยนแปลงการอ้างอิงลูกขวาของพ่อมันไปยังลูกซ้ายของมัน และเปลี่ยนแปลงการอ้างอิงพ่อของลูกซ้ายของมันไปยังพ่อของมัน

#### คำถาม
จงอธิบายว่าทำไมผู้สืบตำแหน่งมีลูกไม่เกิน 1 ปม

ในส่วนของการหาผู้สืบตำแหน่งจะทำการหาโดยการอ้างอิงของปม ดังนั้นจึงทำการสร้างเมธอดช่วยเหลือชื่อว่า findSuccessor ของ TreeNode โดยอาศัยเงื่อนไข
1. ถ้าปมที่ต้องการลบมีลูกขวา ผู้สืบตำแหน่งคือปมที่มีค่าน้อยทีสุดในต้นไม้ย่อยขวา
2. ถ้าไม่มีลูกขวาและเป็นตัวมันเองเป็นลูกซ้าย ผู้สืบตำแหน่งคือพ่อของมัน
3. ถ้าเป็นลูกขวาและตัวมันเองไม่มีลูกขวา ผู้สืบตำแหน่งคือผู้สืบตำแหน่งของพ่อของมันที่ไม่ใช่ตัวมันเอง

มีเมธอดช่วยเหลือคือ findMin เพื่อหาปมที่มีค่าน้อยที่สุดในต้นไม้ย่อย ซึ่งด้วยคุณสมบัติของต้นไม้ค้นหาแบบทวิภาคแล้วก็คือปมที่อยู่ซ้ายที่สุดได้จากการไล่ปมลูกซ้ายไปเรื่อยๆ และมีเมธอดช่วยเหลืออีกหนึ่งตัวคือ spliceOut มีหน้าที่ในการนำผู้สืบตำแหน่งออกจากต้นไม้ด้วยการจัดการการอ้างอิงให้ถูกต้อง สรุปเป็นรหัสได้ทั้งหมดดังนี้

```python
class TreeNode:
    def __init__(self,key,val,left=None,right=None,parent=None):
        self.key = key
        self.payload = val
        self.leftChild = left
        self.rightChild = right
        self.parent = parent

    def hasLeftChild(self):
        return self.leftChild

    def hasRightChild(self):
        return self.rightChild

    def isLeftChild(self):
        return self.parent and self.parent.leftChild == self

    def isRightChild(self):
        return self.parent and self.parent.rightChild == self

    def isRoot(self):
        return not self.parent

    def isLeaf(self):
        return not (self.rightChild or self.leftChild)

    def hasAnyChildren(self):
        return self.rightChild or self.leftChild

    def hasBothChildren(self):
        return self.rightChild and self.leftChild

    def replaceNodeData(self,key,value,lc,rc):
        self.key = key
        self.payload = value
        self.leftChild = lc
        self.rightChild = rc
        if self.hasLeftChild():
            self.leftChild.parent = self
        if self.hasRightChild():
            self.rightChild.parent = self
            
    def spliceOut(self):
       if self.isLeaf():
           if self.isLeftChild():
                  self.parent.leftChild = None
           else:
                  self.parent.rightChild = None
       elif self.hasAnyChildren():
           if self.hasLeftChild():
                  if self.isLeftChild():
                     self.parent.leftChild = self.leftChild
                  else:
                     self.parent.rightChild = self.leftChild
                  self.leftChild.parent = self.parent
           else:
                  if self.isLeftChild():
                     self.parent.leftChild = self.rightChild
                  else:
                     self.parent.rightChild = self.rightChild
                  self.rightChild.parent = self.parent

    def findSuccessor(self):
      succ = None
      if self.hasRightChild():
          succ = self.rightChild.findMin()
      else:
          if self.parent:
                 if self.isLeftChild():
                     succ = self.parent
                 else:
                     self.parent.rightChild = None
                     succ = self.parent.findSuccessor()
                     self.parent.rightChild = self
      return succ

    def findMin(self):
      current = self
      while current.hasLeftChild():
          current = current.leftChild
      return current


class BST:

    def __init__(self):
        self.root = None
        self.size = 0

    def length(self):
        return self.size

    def __len__(self):
        return self.size

    def put(self,key,val):
        if self.root:
            self._put(key,val,self.root)
        else:
            self.root = TreeNode(key,val)
        self.size = self.size + 1

    def _put(self,key,val,currentNode):
        if key < currentNode.key:
            if currentNode.hasLeftChild():
                   self._put(key,val,currentNode.leftChild)
            else:
                   currentNode.leftChild = TreeNode(key,val,parent=currentNode)
        else:
            if currentNode.hasRightChild():
                   self._put(key,val,currentNode.rightChild)
            else:
                   currentNode.rightChild = TreeNode(key,val,parent=currentNode)

    def __setitem__(self,k,v):
       self.put(k,v)

    def get(self,key):
       if self.root:
           res = self._get(key,self.root)
           if res:
                  return res.payload
           else:
                  return None
       else:
           return None

    def _get(self,key,currentNode):
       if not currentNode:
           return None
       elif currentNode.key == key:
           return currentNode
       elif key < currentNode.key:
           return self._get(key,currentNode.leftChild)
       else:
           return self._get(key,currentNode.rightChild)

    def __getitem__(self,key):
       return self.get(key)

    def __contains__(self,key):
       if self._get(key,self.root):
           return True
       else:
           return False

    def delete(self,key):
      if self.size > 1:
         nodeToRemove = self._get(key,self.root)
         if nodeToRemove:
             self.remove(nodeToRemove)
             self.size = self.size-1
         else:
             raise KeyError('Error, key not in tree')
      elif self.size == 1 and self.root.key == key:
         self.root = None
         self.size = self.size - 1
      else:
         raise KeyError('Error, key not in tree')

    def __delitem__(self,key):
       self.delete(key)

    def remove(self,currentNode):
         if currentNode.isLeaf(): #leaf
           if currentNode == currentNode.parent.leftChild:
               currentNode.parent.leftChild = None
           else:
               currentNode.parent.rightChild = None
         elif currentNode.hasBothChildren(): #interior
           succ = currentNode.findSuccessor()
           succ.spliceOut()
           currentNode.key = succ.key
           currentNode.payload = succ.payload

         else: # this node has one child
           if currentNode.hasLeftChild():
             if currentNode.isLeftChild():
                 currentNode.leftChild.parent = currentNode.parent
                 currentNode.parent.leftChild = currentNode.leftChild
             elif currentNode.isRightChild():
                 currentNode.leftChild.parent = currentNode.parent
                 currentNode.parent.rightChild = currentNode.leftChild
             else:
                 currentNode.replaceNodeData(currentNode.leftChild.key,
                                    currentNode.leftChild.payload,
                                    currentNode.leftChild.leftChild,
                                    currentNode.leftChild.rightChild)
           else:
             if currentNode.isLeftChild():
                 currentNode.rightChild.parent = currentNode.parent
                 currentNode.parent.leftChild = currentNode.rightChild
             elif currentNode.isRightChild():
                 currentNode.rightChild.parent = currentNode.parent
                 currentNode.parent.rightChild = currentNode.rightChild
             else:
                 currentNode.replaceNodeData(currentNode.rightChild.key,
                                    currentNode.rightChild.payload,
                                    currentNode.rightChild.leftChild,
                                    currentNode.rightChild.rightChild)

mytree = BST()
mytree[3]="red"
mytree[4]="blue"
mytree[6]="yellow"
mytree[2]="at"

del mytree[3] 

print(mytree[6])
print(mytree[2])
```

### 3.3 การวิเคราะห์ BST
เมื่อเราดูจากเมธอดที่ต้องค้นหาจะเห็นว่ามีแบ่งปัญหาย่อยด้วยการค้นหาต้นไม้ย่อยซ้ายและขวา ซึ่งจำนวนครั้งในการแบ่งนั้นจะถูกกำหนดด้วยความสูงของต้นไม้ ที่เป็นเช่นนี้เพราะการเปรียบเทียบจะเกิดที่ทุกระดับชั้น ดังนั้นเราจึงคาดหวังให้ต้นไม้นั้นมีความสูงน้อยๆ หรือนั่นคือต้นไม้มีการกระจายตัวอย่างสมดุล คือ ต้นไม้ย่อยด้านขวามีขนาดเท่ากับต้นไม่ย่อยด้านซ้าย (หรือต่างกันไม่เกิน 1) เมื่อต้นไม้มีความสมดุลไม่ว่าจะทำกระบวนการใดก็จะใช้เวลา $O(\log n)$ แต่จากการสร้าง BST เราไม่มีกระบวนการใดเลยที่รับประกันว่าต้นไม้ที่ได้จะสมดุล ซึ่งอาจจะเจอในกรณีที่แย่ที่สุดคือ ลำดับการนำเข้าข้อมูลเรียงลำดับไว้อยู่แล้วทำให้เกิดต้นไม้ใน [รูปที่ 5](#figure_05) (จาก [interactivepython](http://interactivepython.org/runestone/static/pythonds/Trees/SearchTreeImplementation.html)) ซึ่งทุกกระบวนการที่ต้องมีการค้นหาจะกลายเป็นใช้เวลา $O(n)$ ทั้งหมด
